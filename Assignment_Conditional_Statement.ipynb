```python
# Question 1 : Write a program that takes the current temperature and provides clothing advice

temp = float(input("Enter current temperature (in °C): "))

if temp < 10:
    print("It's very cold. Wear a jacket or sweater.")
elif temp < 20:
    print("It's cool. Wear a light jacket.")
elif temp < 30:
    print("Weather is warm. T-shirt is fine.")
else:
    print("It's hot. Wear light and breathable clothes.")
```

```python
# Question 2 : Create a program that determines if an applicant qualifies for a loan based on their
# salary and years at their current job.


salary = float(input("Enter your monthly salary: "))
years = int(input("Enter years at current job: "))


if salary >= 30000 and years >= 2:
    print("You qualify for the loan.")
else:
    print("You do not qualify for the loan.")


```

```python
# Question 3 : Develop a script that assigns grades to students based on their score.


score = int(input("Enter student score: "))


if score >= 90:
    print("Grade: A")
elif score >= 75:
    print("Grade: B")
elif score >= 60:
    print("Grade: C")
elif score >= 40:
    print("Grade: D")
else:
    print("Grade: Fail")

```

```python
#Question 4 :  Write a Python script that asks for the number of items and the price per item,
# then calculates and prints the total cost with a 10% discount for 10 or more items, or a 20% discount for 20 or more items.


items = int(input("Enter number of items: "))
price = float(input("Enter price per item: "))


total = items * price


if items >= 20:
    discount = total * 0.20
elif items >= 10:
    discount = total * 0.10
else:
    discount = 0


final_amount = total - discount


print("Total cost:", total)
print("Discount:", discount)
print("Amount to pay:", final_amount)


```
